# quran verses length-statistical research
in this notebook first we want to calculate length of every verse in holy quran and second count the verses which are same in the aspect of length.

```python
import pandas as pd
import json
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
quran_raw = pd.read_csv('quran-simple-clean.txt', header=None)
```

```python
quran_raw.head()
```

```python
quran = []
for verse in list(quran_raw[0]):
    quran.append(len(verse.replace(" ","")))
print(quran)
```

```python
len(quran)
```

```python
count = {}
def add_count(n):
    if n not in count.keys():
        count[n] = 1
    else:
        count[n] += 1
```

```python
for verse_len in quran:
    add_count(verse_len)
```

```python
print(json.dumps(count, indent=1, sort_keys=True))
```

```python
figure(num=None, figsize=(50, 20), dpi=80, facecolor='gray', edgecolor='k')
plt.bar(range(len(count)), list(sorted(count, key=count.get)), align='center')
plt.xticks(range(len(count)), [(k, count[k]) for k in sorted(count, key=count.get)])
plt.show()
```

find the verses with specific length:

```python
def find_verse_by_length(n):
    verses_no = []
    for i in range(len(quran)):
        if n==quran[i]:
            verses_no.append(i)
    for i in range(len(quran_raw[0])):
        if i in verses_no:
            print(quran_raw[0][i])
            verses_no.remove(i)
            if len(verses_no) == 0:
                return
```
